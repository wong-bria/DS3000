## Question 3: 
Evaluate the data for missing or invalid values. Then prepare the data, and ensure that you handle missing and invalid values (if they exist).

```python
# Check for missing values
df.isnull().sum()

# Check for invalid values:
# For instance, negative values in columns that should have positive values
invalid_values = {
    "Pregnancies": df["Pregnancies"] < 0,
    "Glucose": df["Glucose"] <= 0,  
    "BloodPressure": df["BloodPressure"] <= 0,  
    "SkinThickness": df["SkinThickness"] <= 0,  
    "Insulin": df["Insulin"] < 0,  
    "BMI": df["BMI"] <= 0,  
    "DiabetesPedigreeFunction": df["DiabetesPedigreeFunction"] < 0,  
    "Age": df["Age"] < 0,  
    "Outcome": (df["Outcome"] < 0) | (df["Outcome"] > 1)
}

# Check invalid values in each column
for col, condition in invalid_values.items():
    print(f"Invalid values in {col}: {df[condition].shape[0]}")

# Convert to categorical
df['Outcome'] = df['Outcome'].astype('category')
```

In glucose, there are 5 invalid data, 35 in blood pressure, 227 in skin thickness, and 11 in BMI.

For glucose, the invalid data are glucose of 0, which are invalid because it would mean the person probably
dead as that low of a glucose level is fatal.
For blood pressure, the invalid data are blood pressure of 0, which are invalid because it would mean they don't
have any blood pumping, so the person would be dead.
For skin thickness, the invalid data are skin thickness of 0, which are invalid because it would mean they have
no skin, which isn't possible.
For BMI, the invalid data are BMI of 0, which are invalid because it would mean there is no weight, which isn't
possible.

In all these cases, since they are less than 50% of the data, I performed imputation by replacing the invalid
data with the median of each feature.

I also converted Outcome to type category since it is only either 0 or 1.

```python
# Replace invalid values in 'Glucose' with median
df['Glucose'] = df['Glucose'].replace(0, df['Glucose'].median())

# Replace invalid values in 'BloodPressure' with median
df['BloodPressure'] = df['BloodPressure'].replace(0, df['BloodPressure'].median())

# Replace invalid values in 'SkinThickness' with median
df['SkinThickness'] = df['SkinThickness'].replace(0, df['SkinThickness'].median())

# Replace invalid values in 'BMI' with median
df['BMI'] = df['BMI'].replace(0, df['BMI'].median())
```

## Question 5: 
Build the model that predicts the target variable. Ensure that you:

- Use GridSearchCV to tune at least two hyperparameters.**
- Perform cross validation.

**You'll need to read the documentation of the SVM algorithm to identify the hyperparameters.

```python
from sklearn.svm import SVC

# Instantiate the SVM model
svm = SVC()

# Train the model
svm.fit(X_train, y_train)

# Make predictions
y_pred = svm.predict(X_test)

#evaluate the prediction accuracy
print('The accuracy of the model is: {}'.format(svm.score(X_test, y_test)))
```

The accuracy of the SVM model is 73.59%, so it is doing reasonably well on the test set, but still
has room for improvement.

```python
from sklearn.model_selection import GridSearchCV
import numpy as np

# Define the hyperparameters for tuning
params = {
    'C': [0.1, 1, 10, 100, 500],
    'gamma': [0.001, 0.01, 0.1, 1, 10]
}

#initialize Grid Search with Cross Validation
svm_grid = GridSearchCV(estimator = svm, param_grid = params, cv=5)
svm_grid.fit(X_train, y_train)

print(svm_grid.best_params_) #print the best parameters
print(svm_grid.best_score_)  #print the best score

# Evaluate the best model on the test set
test_accuracy = svm_grid.score(X_test, y_test)
print("Test Set Accuracy:", test_accuracy)
```
